```python
# importing libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# Import Data

```python
# importing data
kc_house_data = pd.read_csv(r'/Users/madelinegamache/Downloads/kc_house_data.csv', index_col = False)
```

```python
# Fixing date column
kc_house_data.date = pd.to_datetime(kc_house_data.date)
```

```python
# Checking for duplicates
kc_house_data.duplicated()
```

```python
# Splitting date into 'Day', 'Month', and 'Year'
kc_house_data['year'] = kc_house_data['date'].dt.year
kc_house_data['month'] = kc_house_data['date'].dt.month
kc_house_data['day'] = kc_house_data['date'].dt.day
# Dropping date column
kc_house_data = kc_house_data.drop("date",axis=1)
```

```python
# Check output
kc_house_data.head()
```

```python
# checking for null values
kc_house_data.isnull().sum()
```

```python
# Dropping id column since it will not be used in analysis
kc_house_data = kc_house_data.drop(['id'],axis=1)
```

```python
kc_house_data.shape
```

```python
# Checking for mixed type columns
for col in kc_house_data.columns.tolist():
  weird = (kc_house_data[[col]].applymap(type) != kc_house_data[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (kc_house_data[weird]) > 0:
    print (col)
```

```python
kc_house_data.dtypes
```

```python
# identifying outliers

# Select columns for box plot
cols = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement']

# Create subplots
fig, axes = plt.subplots(1, len(cols), figsize=(20, 5))

# Generate boxplots for each column
for i, col in enumerate(cols):
    sns.boxplot(y=col, data=kc_house_data, ax=axes[i])
    axes[i].set_title(col, fontsize=14)
    axes[i].set_ylabel('')
    
# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Remove outliers for bedrooms column
bedrooms_upper_limit = 12
kc_house_data = kc_house_data[kc_house_data['bedrooms'] <= bedrooms_upper_limit]
```

```python
# Remove outliers for sqft_living column
sqft_living_upper_limit = 10000
kc_house_data = kc_house_data[kc_house_data['sqft_living'] <= sqft_living_upper_limit]
```

```python
# Remove outliers for sqft_basement column
sqft_basement_upper_limit = 3500
kc_house_data = kc_house_data[kc_house_data['sqft_basement'] <= sqft_basement_upper_limit]
```

```python
# Remove outliers for sqft_lot column
sqft_lot_upper_limit = 65000
kc_house_data = kc_house_data[kc_house_data['sqft_lot'] <= sqft_lot_upper_limit]
```

I left outliers in place for price, bathrooms, and sqft_above as they all seemed reasonable.
