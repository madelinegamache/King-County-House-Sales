```python
# Create a correlation matrix for numerical variables
corr_matrix = kc_house_data.corr()

# Visualize the correlation matrix using a heatmap
fig, ax = plt.subplots(figsize=(14, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
```

The following features show a positive correlation with the target variable "price" :
"bathrooms", "sqft_living", "grade", "sqft_living15", and "sqft_above".

"zipcode", "yr_built", and "condition" do not have a significant correlation with "price" and are not adding much value to the dataset. Therefore, we drop these features.

```python
# dropping unneccessary columns for analysis 
kc_house_data = kc_house_data.drop(["zipcode", "yr_built", "condition"], axis = 1)
```
